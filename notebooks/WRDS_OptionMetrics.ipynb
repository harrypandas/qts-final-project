````python
crypto_df = db.raw_sql(
"""
SELECT * 
FROM optionm_all.opprcd2023
WHERE
          (POSITION('BTC' in symbol)=1 OR POSITION('ETH' in symbol)=1)
    AND   date='2023-06-14'
"""
    , )
crypto_df.shape
````

````python
crypto_df[['Underlying', 'Desc']]=crypto_df['symbol'].str.split(" ", n=1, expand=True)
````

````python
crypto_df[['exdate', 'Underlying', 'cp_flag', 'strike_price', 
           'best_bid', 'best_offer',  'delta', 'gamma', 'vega', 
           'volume', 'open_interest', 'impl_volatility',]]
````
